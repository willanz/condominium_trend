```python
from sklearn.linear_model import ElasticNet, Lasso
from sklearn.ensemble import RandomForestRegressor,  GradientBoostingRegressor,BaggingRegressor,AdaBoostRegressor
from sklearn.svm import LinearSVR
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.feature_selection import RFE, SelectFromModel
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_squared_error
import warnings
warnings.filterwarnings('ignore')
regressors = {
    "lasso":Lasso(),
    "enet":ElasticNet(),
    "rfreg":RandomForestRegressor(),
    "lgbreg":LGBMRegressor(),
    "xgbreg":XGBRegressor(objective ='reg:squarederror'),
    "adbreg":AdaBoostRegressor(base_estimator = XGBRegressor(objective = 'reg:squarederror')),
    "breg":BaggingRegressor(base_estimator = XGBRegressor(objective = 'reg:squarederror'))
}
no_regressors = len(regressors.keys())
#models_to_train = [lasso,ridge,enet,rfreg,lsvr,gbreg,xgbreg]
```

```python
# Defining the hyperparameters to optimize
grids = dict(lgrid = { # Lasso
        'rgs__alpha':[0.1,0.11,0.12,0.09],
        'rgs__normalize':[True,False],
        'rgs__precompute':[True,False],
        'rgs__max_iter':[500,1000,1500]},
    egrid = { #Elastic Net
        'rgs__alpha':[0.001,0.0001,0.0002,0.00001],
        'rgs__normalize':[True, False],
        'rgs__tol':[0.00001,0.000001,0.000001],
        'rgs__selection':['cyclic','random']},
    rfgrid = { #Random Forest
        'rgs__max_depth':[9, 11, 13, 15], 
        'rgs__n_estimators':[250,300,350,400],
        'rgs__max_features':[8,10,12,14,16]},
    lgbgrid = { #Light Gradient Boosting
        'rgs__learning_rate':[0.1,0.01,0.002,0.001],
        'rgs__n_estimators':[500,600,700,800],
        'rgs__max_depth':[3,5,6,7]},
    xgbgrid = { #XGBoost
        'rgs__colsample_bytree': [0.3, 0.2, 0.1],
        'rgs__n_estimators': [100,125,150,200],
        'rgs__max_depth': [4, 5,6 ,7 ],
        'rgs__eta': [0.001,0.01,0.1]},
    adbgrid ={ #Adaboost
        'rgs__n_estimators':[50,100,150],
        'rgs__learning_rate':[0.1,0.01,0.02,0.011],
        'rgs__loss':['linear','square']},
    bgrid = { #Bagging
        'rgs__n_estimators':[12,24,48,50],
        'rgs__max_features':[20,25,30,35],
        'rgs__bootstrap_features':[True,False]},
) 
#LGBMRegressor()
```

```python
def batch_regress(X_train, y_train, X_test, y_test, verbose = True):
    df_results = pd.DataFrame(data=np.zeros(shape=(no_regressors,3)), columns = ['regressor','RMSE','R2'])
    count = 0
    #count2 = 0
    scorers = {'R2': 'r2','nmse': 'neg_mean_squared_error'}
    for key, regressor in regressors.items():
        for key2, grid in grids.items():
            scaler = StandardScaler()
            #selector = RFE(regressor, n_features_to_select = 20, step = 2)
            pipe = Pipeline([('scale',scaler), ('rgs',regressor)])#scale and apply the regressor
            if key is 'lasso' and key2 is 'lgrid':
                print(key,key2)
                rand_search = GridSearchCV(pipe, param_grid=grid, scoring=scorers,cv=5, verbose=1, n_jobs=-1, refit='nmse')
                rand_search.fit(X_train, y_train)
                #RMSE = np.sqrt(np.abs(rand_search.cv_results_['mean_test_nmse']))
                #R2 = np.sqrt(np.abs(rand_search.cv_results_['mean_test_R2']))
                count +=1
                print("Best parameters found: ", rand_search.best_params_)
                print("Lowest Train RMSE found: ", np.sqrt(np.abs(rand_search.best_score_)))
                #print("Lowest Train RMSE found: ", np.sqrt(np.abs(rand_search.cv_results_['mean_test_nmse'])))
                #print("Lowest Train R2 found: ", np.sqrt(np.abs(rand_search.cv_results_['mean_test_R2'])))
                preds = rand_search.best_estimator_.predict(X_test)
                test_rmse = np.sqrt(mean_squared_error(y_test, preds))
                test_r2 = r2_score(y_test, preds)
                df_results.loc[count,'regressor'] = key
                df_results.loc[count,'RMSE'] = test_rmse
                df_results.loc[count,'R2'] = test_r2
                print("Test RMSE found: ", test_rmse)
                print("Test R2 found: ", test_r2)
            
            elif key is 'enet' and key2 is 'egrid':
                print(key,key2)
                rand_search = GridSearchCV(pipe, param_grid=grid, scoring=scorers,cv=5, verbose=1, n_jobs=-1, refit='nmse')
                rand_search.fit(X_train, y_train)
                #RMSE = np.sqrt(np.abs(rand_search.cv_results_['mean_test_nmse']))
                #R2 = np.sqrt(np.abs(rand_search.cv_results_['mean_test_R2']))
                count +=1
                print("Best parameters found: ", rand_search.best_params_)
                print("Lowest Train RMSE found: ", np.sqrt(np.abs(rand_search.best_score_)))
                #print("Lowest Train RMSE found: ", np.sqrt(np.abs(rand_search.cv_results_['mean_test_nmse'])))
                #print("Lowest Train R2 found: ", np.sqrt(np.abs(rand_search.cv_results_['mean_test_R2'])))
                preds = rand_search.best_estimator_.predict(X_test)
                test_rmse = np.sqrt(mean_squared_error(y_test, preds))
                test_r2 = r2_score(y_test, preds)
                df_results.loc[count,'regressor'] = key
                df_results.loc[count,'RMSE'] = test_rmse
                df_results.loc[count,'R2'] = test_r2
                print("Test RMSE found: ", test_rmse)
                print("Test R2 found: ", test_r2)
            
            elif key is 'rfreg' and key2 is 'rfgrid':
                print(key,key2)
                rand_search = GridSearchCV(pipe, param_grid=grid, scoring=scorers,cv=5, verbose=1, n_jobs=-1, refit='nmse')
                rand_search.fit(X_train, y_train)
                #RMSE = np.sqrt(np.abs(rand_search.best_score_))
                #df_results.loc[count,'regressor'] = key
                #df_results.loc[count,'RMSE'] = RMSE
                count +=1
                print("Best parameters found: ", rand_search.best_params_)
                print("Lowest Train RMSE found: ", np.sqrt(np.abs(rand_search.best_score_)))
                preds = rand_search.best_estimator_.predict(X_test)
                test_rmse = np.sqrt(mean_squared_error(y_test, preds))
                test_r2 = r2_score(y_test, preds)
                df_results.loc[count,'regressor'] = key
                df_results.loc[count,'RMSE'] = test_rmse
                df_results.loc[count,'R2'] = test_r2
                print("Test RMSE found: ", test_rmse)
                print("Test R2 found: ", test_r2)
            
            elif key is 'lgbreg' and key2 is 'lgbgrid':
                print(key,key2)
                rand_search = GridSearchCV(pipe, param_grid=grid, scoring=scorers,cv=5, verbose=1, n_jobs=-1, refit='nmse')
                rand_search.fit(X_train, y_train)
                #RMSE = np.sqrt(np.abs(rand_search.best_score_))
                #df_results.loc[count,'regressor'] = key
                #df_results.loc[count,'RMSE'] = RMSE
                count +=1
                print("Best parameters found: ", rand_search.best_params_)
                print("Lowest Train RMSE found: ", np.sqrt(np.abs(rand_search.best_score_)))
                preds = rand_search.best_estimator_.predict(X_test)
                test_rmse = np.sqrt(mean_squared_error(y_test, preds))
                test_r2 = r2_score(y_test, preds)
                df_results.loc[count,'regressor'] = key
                df_results.loc[count,'RMSE'] = test_rmse
                df_results.loc[count,'R2'] = test_r2
                print("Test RMSE found: ", test_rmse)
                print("Test R2 found: ", test_r2)
            
            elif key is 'xgbreg' and key2 is 'xgbgrid':
                print(key,key2)           
                rand_search = GridSearchCV(pipe, param_grid=grid, scoring=scorers,cv=5, verbose=1, n_jobs=-1, refit='nmse')
                rand_search.fit(X_train, y_train)
                #RMSE = np.sqrt(np.abs(rand_search.best_score_))
                #df_results.loc[count,'regressor'] = key
                #df_results.loc[count,'RMSE'] = RMSE
                count +=1
                print("Best parameters found: ", rand_search.best_params_)
                print("Lowest Train RMSE found: ", np.sqrt(np.abs(rand_search.best_score_)))
                preds = rand_search.best_estimator_.predict(X_test)
                test_rmse = np.sqrt(mean_squared_error(y_test, preds))
                test_r2 = r2_score(y_test, preds)
                df_results.loc[count,'regressor'] = key
                df_results.loc[count,'RMSE'] = test_rmse
                df_results.loc[count,'R2'] = test_r2
                print("Test RMSE found: ", test_rmse)
                print("Test R2 found: ", test_r2)
            
            elif key is 'adbreg' and key2 is 'adbgrid':
                print(key,key2)
                rand_search = GridSearchCV(pipe, param_grid=grid, scoring=scorers,cv=5, verbose=1, n_jobs=-1,refit='nmse')
                rand_search.fit(X_train, y_train)
                #RMSE = np.sqrt(np.abs(rand_search.best_score_))
                #df_results.loc[count,'regressor'] = key
                #df_results.loc[count,'RMSE'] = RMSE
                count +=1
                print("Best parameters found: ", rand_search.best_params_)
                print("Lowest Train RMSE found: ", np.sqrt(np.abs(rand_search.best_score_)))
                preds = rand_search.best_estimator_.predict(X_test)
                test_rmse = np.sqrt(mean_squared_error(y_test, preds))
                test_r2 = r2_score(y_test, preds)
                df_results.loc[count,'regressor'] = key
                df_results.loc[count,'RMSE'] = test_rmse
                df_results.loc[count,'R2'] = test_r2
                print("Test RMSE found: ", test_rmse)
                print("Test R2 found: ", test_r2)
            
            elif key is 'breg' and key2 is 'bgrid':
                print(key,key2)           
                rand_search = GridSearchCV(pipe, param_grid=grid, scoring=scorers,cv=5, verbose=1, n_jobs=-1,refit='nmse')
                rand_search.fit(X_train, y_train)
                #RMSE = np.sqrt(np.abs(rand_search.best_score_))
                #df_results.loc[count,'regressor'] = key
                #df_results.loc[count,'RMSE'] = RMSE
                count +=1
                print("Best parameters found: ", rand_search.best_params_)
                print("Lowest Train RMSE found: ", np.sqrt(np.abs(rand_search.best_score_)))
                preds = rand_search.best_estimator_.predict(X_test)
                test_rmse = np.sqrt(mean_squared_error(y_test, preds))
                test_r2 = r2_score(y_test, preds)
                df_results.loc[count,'regressor'] = key
                df_results.loc[count,'RMSE'] = test_rmse
                df_results.loc[count,'R2'] = test_r2
                print("Test RMSE found: ", test_rmse)
                print("Test R2 found: ", test_r2)
            else:
                pass
    return df_results
```
